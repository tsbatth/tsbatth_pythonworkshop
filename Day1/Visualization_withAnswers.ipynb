```python
for el in range(10):
    fig, (ax1, ax2) = plt.subplots(1, 2)
    ax1.hist(np.random.normal(size=1000), bins=100, cumulative=True, alpha=.5);
    ax1.set(xlabel='dsafaf', ylabel='ddd', title='fdafaf')
    ax2.hist(np.random.normal(size=1000), bins=100, cumulative=True, alpha=.5);
    ax2.set_ylabel('ffff')
    fig.savefig('afda{}.png'.format(el))
```

## Scatter Plots
  - x label: "Height"
  - y label: "Weight"
  - title: "Relationship between Height and Weight in Adults"  
  
**Note**: Fake data looks more believable when it has some variance (i.e. "noise") to it.

```python
x = np.random.normal(2, .2, size=30)
noise = np.random.normal(0, 5, size=30)
y = 50 * x + noise
plt.scatter(x, y, c='red', label='Adults');

noise2 = np.random.normal(0, 5, size=30)

y2 = 60 * x + noise2
plt.scatter(x, y2, c='green', label='Children')

plt.legend()
```
