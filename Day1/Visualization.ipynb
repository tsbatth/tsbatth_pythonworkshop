## Histograms

Make three histograms:

**a uniform distribution:**

```python
fig = plt.figure()
plt.hist(np.random.uniform(size=1000), bins=50);
```

**a normal distribution:**

  - x label: "Height of Children in Class"
  - y label: "Count"

```python
fig2 = plt.figure()
plt.hist(np.random.normal(size=1000), bins=50);
plt.xlabel('Height of Children in Class')
plt.ylabel('Count')
```

**a poisson distribution:**

  - x label: "Inter-Protein Arrival Times"
  - y label: "Frequency"
  - title: "Peptide Time Intervals between LC Elutions"

```python
fig2 = plt.figure()
plt.hist(np.random.poisson(size=1000), bins=50);
plt.xlabel('Inter-Protein Arrival Times')
plt.ylabel('Frequency')
plt.title('Peptide Time Intervals between LC Elutions')
```

## Scatter Plots
  - x label: "Height"
  - y label: "Weight"
  - title: "Relationship between Height and Weight in Adults"  
  
**Note**: Fake data looks more believable when it has some variance (i.e. "noise") to it.

```python
fig3 = plt.figure()
plt.scatter(x=np.random.normal(2,.2, size=50), y=np.random.normal(5, 5, size=50))
plt.xlabel('Height')
plt.ylabel('Weight')
```
